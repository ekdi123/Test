```python
t=np.arange(0,12,0.01)
y1 = np.sin(t)
y2 = np.cos(t)
plt.grid()
plt.plot(t,y1,'y--')
plt.plot(t,y2,'r-')

plt.xlabel('Time')
plt.ylabel('Amp')
plt.title('SinCos')
plt.legend(['Sin','Cos'])
```
